```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# Import data from simulated_clustering.ipynb
# Using %run magic command to execute the clustering notebook and inherit its variables
%run simulated_clustering.ipynb

# Verify that df and customer_df are now available in this notebook
print("Data imported from simulated_clustering.ipynb")
print(f"df shape: {df.shape}")
print(f"customer_df shape: {customer_df.shape}")
print("\ndf columns:", df.columns.tolist())
print("\ncustomer_df columns:", customer_df.columns.tolist())

# Get the latest date from the data for analysis
latest_date = pd.to_datetime(df['Date']).max()
print(f"\nLatest date in data: {latest_date}")

```

```python
# Calculate today's date for reference
from datetime import datetime, timedelta

today = latest_date#datetime.now()
today_date = today.date()

# Define time thresholds
three_months_ago = today - timedelta(days=90)  # approximately 1 quarter
one_year_ago = today - timedelta(days=365)

print(f"Today's date: {today_date}")
print(f"Three months ago: {three_months_ago.date()}")
print(f"One year ago: {one_year_ago.date()}")

# Convert date columns to datetime if they aren't already
customer_df['first_purchase_date'] = pd.to_datetime(customer_df['first_purchase_date'])
customer_df['last_purchase_date'] = pd.to_datetime(customer_df['last_purchase_date'])

# Drop existing columns if they exist to avoid merge conflicts
columns_to_drop = ['total_quarters_active', 'quarters_list', 'recent_quarters', 'quarters_active_last_2']
customer_df = customer_df.drop(columns=[col for col in columns_to_drop if col in customer_df.columns])

# Since we need continuous purchasing data but only have aggregated customer data,
# we'll need to work with what we have in customer_df and supplement with original df

# First, let's analyze purchasing patterns using the original transaction data
print("\n=== Analyzing Purchase Continuity from Original Data ===")

# Check what columns are available in df
print("Available columns in df:", df.columns.tolist())


# Create a more detailed customer analysis using the original df
# Group by customer and quarter to see purchase continuity
df_analysis = df.copy()
df_analysis['date'] = pd.to_datetime(df_analysis['Date'])
df_analysis['quarter_year'] = df_analysis['date'].dt.to_period('Q')


# Get customer purchase activity by quarter using available columns
customer_quarterly_activity = df_analysis.groupby(['Customer_Group', 'quarter_year']).agg({
    'Amount': 'sum',
    'Doc_No': 'nunique',
    'Date': 'count'
}).reset_index()

# Rename columns for consistency
customer_quarterly_activity = customer_quarterly_activity.rename(columns={
    'Customer_Group': 'customer_group',  # Convert to lowercase to match customer_df
    'Amount': 'monetary_sum',
    'Doc_No': 'order_count',
    'Date': 'transaction_count'
})

# Get the last 8 quarters from today for analysis  Note: this is not fiscal quarter, it is calendar quarter based on today's date
current_quarter = pd.Period(today, freq='Q')
last_8_quarters = [current_quarter - i for i in range(8)]
last_2_quarters = [current_quarter - i for i in range(2)]

print(f"Current quarter: {current_quarter}")
print(f"Last 2 quarters for retention analysis: {last_2_quarters}")
print(f"Last 8 quarters for comprehensive view: {last_8_quarters}")

# Analyze customer activity in recent quarters
def analyze_quarterly_activity(customer_group):
    customer_data = customer_quarterly_activity[
        customer_quarterly_activity['customer_group'] == customer_group
    ]
    
    # Check activity in last 2 quarters
    recent_activity = customer_data[
        customer_data['quarter_year'].isin(last_2_quarters)
    ]
    
    return {
        'quarters_active_last_2': len(recent_activity),
        'total_quarters_active': len(customer_data),
        'quarters_list': customer_data['quarter_year'].tolist(),
        'recent_quarters': recent_activity['quarter_year'].tolist()
    }



# Apply quarterly analysis to each customer
print("\nAnalyzing quarterly activity patterns...")
quarterly_analysis = {}
for customer in customer_df['customer_group'].unique():
    if pd.notna(customer):  # Skip NaN values
        quarterly_analysis[customer] = analyze_quarterly_activity(customer)


# Convert to DataFrame for easier analysis
quarterly_df = pd.DataFrame.from_dict(quarterly_analysis, orient='index')
quarterly_df = quarterly_df.reset_index().rename(columns={'index': 'customer_group'})


# Merge with customer_df
customer_df = customer_df.merge(quarterly_df, on='customer_group', how='left')

# Fill NaN values with 0 for customers not found in quarterly analysis
customer_df['quarters_active_last_2'] = customer_df['quarters_active_last_2'].fillna(0)
customer_df['total_quarters_active'] = customer_df['total_quarters_active'].fillna(0)


# Calculate days between last two purchases for Win Back detection
# First, get the second-to-last purchase date for each customer
print("\nCalculating days between purchases for Win Back detection...")

# Get purchase dates for each customer (excluding duplicates, sorted by date)
customer_purchase_dates = df_analysis.groupby('Customer_Group')['date'].apply(
    lambda x: sorted(x.drop_duplicates(), reverse=True)
).to_dict()



def get_days_between_last_purchases(customer_group):
    """Calculate days between last two purchases"""
    if customer_group not in customer_purchase_dates:

        return None
    
    dates = customer_purchase_dates[customer_group]
    if len(dates) < 2:
        return None
    
    # Get last two purchase dates
    last_purchase = dates[0]
    second_last_purchase = dates[1]
    
    return (last_purchase - second_last_purchase).days

# Add days between purchases to customer_df
customer_df['days_between_last_purchases'] = customer_df['customer_group'].apply(get_days_between_last_purchases)

# Initialize comprehensive customer behavior classification with Win Back logic
def classify_customer_behavior_comprehensive(row):
    first_purchase = row['first_purchase_date']
    last_purchase = row['last_purchase_date']
    quarters_active_last_2 = row['quarters_active_last_2'] if pd.notna(row['quarters_active_last_2']) else 0
    days_between_purchases = row['days_between_last_purchases']
    doc_no_unique = row['doc_no_unique'] if pd.notna(row['doc_no_unique']) else 0
    
    # New customers: first purchase within last quarter (3 months)
    if first_purchase >= three_months_ago:
        return 'New'
    
    # Churn customers: last purchase more than 1 year ago
    elif last_purchase <= one_year_ago:
        return 'Churn'
    
    # One Time customers: only one purchase record and not new or churn
    elif doc_no_unique == 1:
        return 'One Time'
    
    # Win Back customers: 
    # - Gap >1 year between last two purchases AND 
    # - Last purchase within 3 months (recent return) AND
    # - Not New (already handled above)
    elif (pd.notna(days_between_purchases) and 
          days_between_purchases > 365 and 
          last_purchase >= three_months_ago):
        return 'Win Back'
    
    # Retained customers: purchased in both of the last 2 quarters
    elif quarters_active_last_2 >= 2:
        return 'Active-2Q'
    
    # At Risk: purchased in only 1 of the last 2 quarters
    elif quarters_active_last_2 == 1:
        return 'Recent-1Q'
    
    # Undefined: no purchases in last 2 quarters but within last year
    else:
        return 'At Risk'

# Apply comprehensive classification
customer_df['behavior_status'] = customer_df.apply(classify_customer_behavior_comprehensive, axis=1)

# Calculate average days between orders for alert system
customer_df['avg_days_between_orders'] = ((customer_df['last_purchase_date'] - customer_df['first_purchase_date']).dt.days / customer_df['doc_no_unique']).round(0)

# Calculate days since last purchase
customer_df['days_since_last_purchase'] = (today - customer_df['last_purchase_date']).dt.days

# Alert flag: if days since last purchase > 1.5 times average days between orders
customer_df['needs_alert'] = (
    (customer_df['behavior_status'] != 'New') & 
    (
        (customer_df['days_since_last_purchase'] > customer_df['avg_days_between_orders'] * 1.5) |
        (customer_df['behavior_status'] == 'Churn')
    )
)

# Display comprehensive summary
print("\n" + "="*60)
print("COMPREHENSIVE CUSTOMER BEHAVIOR ANALYSIS")
print("="*60)

behavior_summary = customer_df['behavior_status'].value_counts()
print("\n1. Customer count by behavior status:")
for status, count in behavior_summary.items():
    percentage = (count / len(customer_df)) * 100
    print(f"   {status}: {count} customers ({percentage:.1f}%)")

print(f"\n2. Alert Summary:")
alert_summary = customer_df['needs_alert'].value_counts()
total_customers = len(customer_df)
need_alert = alert_summary.get(True, 0)
no_alert = alert_summary.get(False, 0)
print(f"   Customers needing alerts: {need_alert} ({(need_alert/total_customers)*100:.1f}%)")
print(f"   Customers not needing alerts: {no_alert} ({(no_alert/total_customers)*100:.1f}%)")

# Detailed analysis by behavior status
print(f"\n3. Detailed Analysis by Behavior Status:")
print("-" * 50)

for status in ['New', 'Active-2Q', 'Recent-1Q', 'At Risk', 'Win Back', 'One Time', 'Churn']:
    status_customers = customer_df[customer_df['behavior_status'] == status]
    if len(status_customers) > 0:
        print(f"\n{status.upper()} CUSTOMERS ({len(status_customers)} customers):")
        print(f"   Average monetary value: ${status_customers['monetary'].mean():,.2f}")
        print(f"   Average frequency (quarters): {status_customers['frequency_q'].mean():.1f}")
        print(f"   Average recency (days): {status_customers['recency'].mean():.0f}")
        print(f"   Customers needing alerts: {status_customers['needs_alert'].sum()} ({(status_customers['needs_alert'].sum()/len(status_customers))*100:.1f}%)")
        
        # Show Win Back specific metrics
        if status == 'Win Back':
            avg_gap = status_customers['days_between_last_purchases'].mean()
            print(f"   Average gap between purchases: {avg_gap:.0f} days ({avg_gap/365:.1f} years)")
        
        # Top customers by monetary value in each category
        top_customers = status_customers.nlargest(3, 'monetary')[['customer_group', 'monetary', 'frequency_q', 'last_purchase_date']]
        if len(top_customers) > 0:
            print(f"   Top customers by revenue:")
            for _, customer in top_customers.iterrows():
                print(f"     â€¢ {customer['customer_group']}: ${customer['monetary']:,.0f} ({customer['frequency_q']:.0f} quarters)")

# Alert analysis by behavior status
print(f"\n4. Alert Analysis by Behavior Status:")
print("-" * 50)
alert_by_status = customer_df[customer_df['needs_alert'] == True]['behavior_status'].value_counts()
for status, count in alert_by_status.items():
    status_total = len(customer_df[customer_df['behavior_status'] == status])
    percentage = (count / status_total) * 100 if status_total > 0 else 0
    print(f"   {status}: {count}/{status_total} customers need alerts ({percentage:.1f}%)")

# Quarterly activity insights
print(f"\n5. Quarterly Activity Insights:")
print("-" * 50)
print(f"   Average quarters active (lifetime): {customer_df['total_quarters_active'].mean():.1f}")
print(f"   Customers active in both last 2 quarters: {len(customer_df[customer_df['quarters_active_last_2'] >= 2])}")
print(f"   Customers active in only 1 of last 2 quarters: {len(customer_df[customer_df['quarters_active_last_2'] == 1])}")
print(f"   Customers with no activity in last 2 quarters: {len(customer_df[customer_df['quarters_active_last_2'] == 0])}")

# Win Back analysis
print(f"\n6. Win Back Customer Analysis:")
print("-" * 50)
win_back_customers = customer_df[customer_df['behavior_status'] == 'Win Back']
if len(win_back_customers) > 0:
    print(f"   Total Win Back customers: {len(win_back_customers)}")
    print(f"   Average revenue per Win Back customer: ${win_back_customers['monetary'].mean():,.2f}")
    print(f"   Total revenue from Win Back customers: ${win_back_customers['monetary'].sum():,.2f}")
    print(f"   Average gap before return: {win_back_customers['days_between_last_purchases'].mean():.0f} days")
    print(f"   Revenue percentage from Win Back: {(win_back_customers['monetary'].sum()/customer_df['monetary'].sum())*100:.1f}%")
else:
    print("   No Win Back customers identified")

# Sample customer details for verification
print(f"\n7. Sample Customer Classifications:")
print("-" * 50)
sample_customers = customer_df[['customer_group', 'first_purchase_date', 'last_purchase_date', 
                               'quarters_active_last_2', 'days_between_last_purchases',
                               'days_since_last_purchase', 'avg_days_between_orders', 
                               'behavior_status', 'needs_alert']].head(10)
print(sample_customers.to_string(index=False))

# Revenue impact analysis
print(f"\n8. Revenue Impact Analysis:")
print("-" * 50)
total_revenue = customer_df['monetary'].sum()
for status in ['New', 'Active-2Q', 'Recent-1Q', 'At Risk', 'Win Back', 'One Time', 'Churn']:
    status_revenue = customer_df[customer_df['behavior_status'] == status]['monetary'].sum()
    percentage = (status_revenue / total_revenue) * 100
    print(f"   {status}: ${status_revenue:,.0f} ({percentage:.1f}% of total revenue)")

print(f"\n   Total Revenue: ${total_revenue:,.0f}")

# Risk assessment
print(f"\n9. Risk Assessment:")
print("-" * 50)
at_risk_revenue = customer_df[customer_df['behavior_status'].isin(['At Risk'])]['monetary'].sum()
at_risk_percentage = (at_risk_revenue / total_revenue) * 100
print(f"   Revenue at risk: ${at_risk_revenue:,.0f} ({at_risk_percentage:.1f}% of total)")

alert_revenue = customer_df[customer_df['needs_alert'] == True]['monetary'].sum()
alert_percentage = (alert_revenue / total_revenue) * 100
print(f"   Revenue from customers needing alerts: ${alert_revenue:,.0f} ({alert_percentage:.1f}% of total)")

# Win Back opportunity assessment
win_back_revenue = customer_df[customer_df['behavior_status'] == 'Win Back']['monetary'].sum()
win_back_percentage = (win_back_revenue / total_revenue) * 100
print(f"   Revenue from Win Back customers: ${win_back_revenue:,.0f} ({win_back_percentage:.1f}% of total)")

print("\n" + "="*60)
print("ANALYSIS COMPLETE")
print("="*60)
```
